```python
df=df[['ORDERLINENUMBER','SALES']]
```

```python
df
```

```python
from sklearn.preprocessing import StandardScaler
model_scale=StandardScaler()
```

```python
scaled_values=model_scale.fit_transform(df.values)
```

```python
wcss=[]
for i in range(1,11):
    model=KMeans(n_clusters=i,init="k-means++")
    model.fit_predict(scaled_values)
    wcss.append(model.inertia_)
import matplotlib.pyplot as plt
plt.plot(range(1,11),wcss,'ro--')
plt.show
```

```python
model=KMeans(n_clusters=7,init="k-means++")
cluster=model.fit_predict(scaled_values)
cluster
```

```python
df['cluster']=cluster
```
